```python
import matplotlib.pyplot as plt
import numpy as np

file_path_deepmvi = '/home/velosog/shared_dir/blackout_deepmvi/electricity/error/mae/MAE_deepmvi.dat'
file_path_dynnamo = '/home/velosog/shared_dir/dynammo/blackout/electricity/error/mae/MAE_dynammo.dat'

data_deepmvi = np.loadtxt(file_path_deepmvi)
data_dynnamo = np.loadtxt(file_path_dynnamo)

x_deepmvi = data_deepmvi[:, 0]
y_deepmvi = data_deepmvi[:, 1]
x_dynnamo = data_dynnamo[:, 0]
y_dynnamo = data_dynnamo[:, 1]

plt.figure(figsize=(4.67, 2.33))
plt.plot(x_deepmvi, y_deepmvi, color='blue', linewidth=2, marker='o', markerfacecolor='none', markeredgecolor='blue', label='DeepMVI')
plt.plot(x_dynnamo, y_dynnamo, color='red', linewidth=2, marker='^', markerfacecolor='none', markeredgecolor='red', label='Dynammo')

plt.scatter(x_deepmvi, y_deepmvi, edgecolors='blue', facecolors='none', s=50, marker='o', zorder=5)
plt.scatter(x_dynnamo, y_dynnamo, edgecolors='red', facecolors='none', s=50, marker='^', zorder=5)

plt.xlabel('Percentage of Missing Values', fontsize=10, color='black')
plt.ylabel('Mean Absolute Error', fontsize=10, color='black')
plt.ylim(0, 1)

plt.tick_params(axis='both', which='major', labelsize=8, colors='black')
plt.tick_params(axis='both', which='minor', labelsize=6, colors='black')
plt.tick_params(axis='x', which='both', top=True, labeltop=False)
plt.tick_params(axis='y', which='both', right=True, labelright=False)

# Set x-axis ticks starting from 10 at 10-unit intervals
x_min, x_max = plt.xlim()
plt.xticks(np.arange(10, x_max + 1, 10))

plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```
